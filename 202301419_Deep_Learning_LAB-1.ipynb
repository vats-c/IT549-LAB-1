```python


!pip install -q pandas numpy scikit-learn matplotlib seaborn wordcloud spacy tqdm tensorflow
!python -m spacy download en_core_web_sm

import os, html, re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm

import spacy
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_selection import chi2
from wordcloud import WordCloud

import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras import layers, models, callbacks, optimizers


SEED = 42
np.random.seed(SEED)
tf.random.set_seed(SEED)


nlp = spacy.load("en_core_web_sm", disable=["ner", "parser"])

#  Load dataset
df = pd.read_csv("fake_reviews_dataset.csv")
print("Loaded:", df.shape)
print("Columns:", df.columns.tolist())

TEXT_COL = "text_"
CATEGORY_COL = "category"
RATING_COL = "rating"

# Create fake/genuine labels from rating (standard approach)
#   fake = 1 if rating <= 2, genuine = 0 if rating >= 4, drop rating == 3
df = df[df[RATING_COL].notna()].copy()
df["label_fake"] = df[RATING_COL].apply(lambda x: 1 if x <= 2 else (0 if x >= 4 else np.nan))
df = df[df["label_fake"].notna()].reset_index(drop=True)
df["label_fake"] = df["label_fake"].astype(int)
print("After creating label_fake:", df.shape)
print(df["label_fake"].value_counts())

# Preprocess text with spaCy (lemmatize + remove stopwords/punct)
def clean_text_spacy(s):
    s = html.unescape(str(s))
    doc = nlp(s.lower())
    tokens = [token.lemma_ for token in doc if token.is_alpha and not token.is_stop]
    return " ".join(tokens)

print("Cleaning text with spaCy (this can take 1-2 minutes)...")
texts = df[TEXT_COL].astype(str).tolist()
cleaned = []
for doc in tqdm(nlp.pipe(texts, batch_size=64, n_process=1), total=len(texts)):
    toks = [token.lemma_ for token in doc if token.is_alpha and not token.is_stop]
    cleaned.append(" ".join(toks))
df["clean_text"] = cleaned

#  Prepare tokenizer + sequences (shared vocabulary for both tasks)
MAX_NUM_WORDS = 20000
MAX_SEQUENCE_LENGTH = 200
EMBEDDING_DIM = 128

tokenizer = Tokenizer(num_words=MAX_NUM_WORDS, oov_token="<OOV>")
tokenizer.fit_on_texts(df["clean_text"].tolist())

sequences = tokenizer.texts_to_sequences(df["clean_text"].tolist())
X_all = pad_sequences(sequences, maxlen=MAX_SEQUENCE_LENGTH, padding="post", truncating="post")

# Multiclass: product category classification
do_multiclass = True
if CATEGORY_COL not in df.columns or df[CATEGORY_COL].nunique() < 2:
    do_multiclass = False
    print("Skipping multiclass: category column missing or insufficient classes.")
else:

    le_cat = LabelEncoder()
    y_cat = le_cat.fit_transform(df[CATEGORY_COL].astype(str))
    print("Category classes:", le_cat.classes_.shape[0])


    X_train_cat, X_test_cat, y_train_cat, y_test_cat = train_test_split(
        X_all, y_cat, test_size=0.2, stratify=y_cat, random_state=SEED
    )


    def build_multiclass_model(num_classes):
        inp = layers.Input(shape=(MAX_SEQUENCE_LENGTH,))
        x = layers.Embedding(input_dim=MAX_NUM_WORDS, output_dim=EMBEDDING_DIM, input_length=MAX_SEQUENCE_LENGTH)(inp)
        x = layers.Conv1D(filters=128, kernel_size=5, activation="relu")(x)
        x = layers.GlobalMaxPool1D()(x)
        x = layers.Dense(64, activation="relu")(x)
        x = layers.Dropout(0.4)(x)
        out = layers.Dense(num_classes, activation="softmax")(x)
        model = models.Model(inputs=inp, outputs=out)
        model.compile(optimizer=optimizers.Adam(1e-3), loss="sparse_categorical_crossentropy", metrics=["accuracy"])
        return model

    num_classes = len(le_cat.classes_)
    model_cat = build_multiclass_model(num_classes)
    model_cat.summary()


    es = callbacks.EarlyStopping(monitor="val_loss", patience=3, restore_best_weights=True)
    chk = callbacks.ModelCheckpoint("model_multiclass.h5", save_best_only=True, monitor="val_loss")


    history_cat = model_cat.fit(
        X_train_cat, y_train_cat,
        validation_split=0.1,
        epochs=12,
        batch_size=64,
        callbacks=[es, chk],
        verbose=2
    )


    loss_cat, acc_cat = model_cat.evaluate(X_test_cat, y_test_cat, verbose=0)
    print(f"Multiclass test loss: {loss_cat:.4f}, acc: {acc_cat:.4f}")


    y_pred_cat = model_cat.predict(X_test_cat, verbose=0)
    y_pred_cat_labels = np.argmax(y_pred_cat, axis=1)

    from sklearn.metrics import classification_report, confusion_matrix
    print("Multiclass classification report:")
    print(classification_report(y_test_cat, y_pred_cat_labels, target_names=le_cat.classes_))

#----- Binary: fake review detection -----

y_fake = df["label_fake"].values
X_train_fake, X_test_fake, y_train_fake, y_test_fake = train_test_split(
    X_all, y_fake, test_size=0.2, stratify=y_fake, random_state=SEED
)

def build_binary_model():
    inp = layers.Input(shape=(MAX_SEQUENCE_LENGTH,))
    x = layers.Embedding(input_dim=MAX_NUM_WORDS, output_dim=EMBEDDING_DIM, input_length=MAX_SEQUENCE_LENGTH)(inp)
    x = layers.Conv1D(filters=128, kernel_size=5, activation="relu")(x)
    x = layers.GlobalMaxPool1D()(x)
    x = layers.Dense(64, activation="relu")(x)
    x = layers.Dropout(0.4)(x)
    out = layers.Dense(1, activation="sigmoid")(x)
    model = models.Model(inputs=inp, outputs=out)
    model.compile(optimizer=optimizers.Adam(1e-3), loss="binary_crossentropy", metrics=["accuracy"])
    return model

model_bin = build_binary_model()
model_bin.summary()

from sklearn.utils import class_weight
class_weights = class_weight.compute_class_weight("balanced", classes=np.unique(y_train_fake), y=y_train_fake)
cw = {i: class_weights[i] for i in range(len(class_weights))}
print("Class weights:", cw)

es_bin = callbacks.EarlyStopping(monitor="val_loss", patience=3, restore_best_weights=True)
chk_bin = callbacks.ModelCheckpoint("model_binary.h5", save_best_only=True, monitor="val_loss")

history_bin = model_bin.fit(
    X_train_fake, y_train_fake,
    validation_split=0.1,
    epochs=12,
    batch_size=64,
    callbacks=[es_bin, chk_bin],
    class_weight=cw,
    verbose=2
)


loss_bin, acc_bin = model_bin.evaluate(X_test_fake, y_test_fake, verbose=0)
print(f"Binary test loss: {loss_bin:.4f}, acc: {acc_bin:.4f}")

y_prob = model_bin.predict(X_test_fake, verbose=0).ravel()
y_pred_bin = (y_prob >= 0.5).astype(int)

from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, precision_recall_curve, auc
print("Fake detection classification report:")
print(classification_report(y_test_fake, y_pred_bin))
try:
    print("ROC AUC:", roc_auc_score(y_test_fake, y_prob))
except:
    pass

cm = confusion_matrix(y_test_fake, y_pred_bin)
plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.title("Confusion Matrix (Fake Detection)")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()

cv = CountVectorizer(ngram_range=(1,2), min_df=5, max_features=10000)
X_counts = cv.fit_transform(df["clean_text"].values)
chi2_scores, pvals = chi2(X_counts, df["label_fake"].values)
terms = np.array(cv.get_feature_names_out())
top_idx = np.argsort(chi2_scores)[-200:][::-1]
top_terms = terms[top_idx]
top_scores = chi2_scores[top_idx]
freqs = {w: float(s) for w, s in zip(top_terms, top_scores)}

wc = WordCloud(width=1200, height=600, background_color="white")
wc.generate_from_frequencies(freqs)
plt.figure(figsize=(14,7))
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.title("Word Cloud: Terms correlated with fake reviews (chi2)")
plt.show()
wc.to_file("fake_wordcloud_chi2.png")
print("Saved fake_wordcloud_chi2.png")

os.makedirs("saved_models", exist_ok=True)
model_bin.save("saved_models/fake_detector_nn.h5")
tokenizer_json = tokenizer.to_json()
with open("saved_models/tokenizer.json", "w", encoding="utf-8") as f:
    f.write(tokenizer_json)
if do_multiclass:
    model_cat.save("saved_models/category_classifier_nn.h5")

print("Saved models and tokenizer to saved_models/")
print("Done.")
```
